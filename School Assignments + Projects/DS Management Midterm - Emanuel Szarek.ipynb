```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DF_GDP = pd.read_csv("GDP_by_Country.csv", sep = ",", skiprows= range(0,4))
```

```python
DF_GDP.head()
```

```python
del DF_GDP['Unnamed: 62']
del DF_GDP['2017']
```

```python
DF_GDPT = DF_GDP.T
DF_GDPT.head()
```

```python
DF_GDPT.columns = DF_GDPT.iloc[0]
DF_GDPT.head()
```

```python
def Country_GDP(Country):
    Country_name = DF_GDPT[Country]
    Country_GDP = Country_name[4:]
    return Country_GDP
```

```python
def Graph_GDP(country1, country2, country3):
    plt.xlabel('Year', fontsize = 18) 
    plt.ylabel('GDP Growth', fontsize = 18) 
    plt.title('GDP Growth', fontsize=18) 
    plt.style.use('seaborn-whitegrid')
    
    Country_GDP(country1).plot(legend = 'Upper left', color ='red', label= country1, figsize=(8,6))
    Country_GDP(country2).plot(legend = 'Upper left', color = 'blue', label = country2)
    Country_GDP(country3).plot(legend = 'Upper left', color = 'black', label = country3)
    plt.legend(loc=2, prop={'size': 15})
    plt.tick_params(axis = 'both', which = 'major', labelsize = 14, length = 10.0)
```

```python
def Country_Mortality(Country):
    Country_name = Dot_DF_MFIT[Country][3:50]
    return Country_name.astype(float)

def Country_Fertility(Country):
    Country_name = Dot_DF_MFIT[Country][50:97]
    return Country_name.astype(float)

def Country_Income(Country):
    Country_name = Dot_DF_MFIT[Country][97:]
    return Country_name.astype(float)

```

```python
def Graph_Mortality(country1, country2, country3): 
    plt.xlabel('Year', fontsize = 18) 
    plt.ylabel('Mortality rate (Per 1,000 live births)', fontsize = 18) 

    plt.title('Mortality Rate of '+ country1 + ", " + country2 + " and " + country3 , fontsize=18) 
    plt.style.use('seaborn-whitegrid')
    
    Country_Mortality(country1).plot(legend = 'Upper right', color ='red', label= country1, figsize=(8,6))
    Country_Mortality(country2).plot(legend = 'Upper right', color = 'blue', label = country2)
    Country_Mortality(country3).plot(legend = 'Upper right', color = 'black', label = country3)
    plt.legend(loc=1, prop={'size': 15})
    plt.tick_params(axis = 'both', which = 'major', labelsize = 14, length = 10)
```

```python
def Graph_Fertility(country1, country2, country3): 
    plt.xlabel('Year', fontsize = 18) 
    plt.ylabel('Total (births per woman)', fontsize = 18) 
  
    plt.title('Fertility Rate of '+ country1 + ", " + country2 + " and " + country3 , fontsize=18) 
    plt.style.use('seaborn-whitegrid')
    
    Country_Fertility(country1).plot(legend = 'Upper right', color ='red', label= country1, figsize=(8,6))
    Country_Fertility(country2).plot(legend = 'Upper right', color = 'blue', label = country2)
    Country_Fertility(country3).plot(legend = 'Upper right', color = 'black', label = country3)
    plt.legend(loc=1, prop={'size': 15})
    plt.tick_params(axis = 'both', which = 'major', labelsize = 14, length = 10)
```

```python
def Graph_Income(country1, country2, country3): 
    plt.xlabel('Year', fontsize = 18) 
    plt.ylabel('US $', fontsize = 18) 
  
    plt.title('Adjusted net national income per capita of '+ country1 + ", " + country2 + " and " + country3 , fontsize=14) 
    plt.style.use('seaborn-whitegrid')
    
    Country_Income(country1).plot(legend = 'Upper right', color ='red', label= country1, figsize=(8,6))
    Country_Income(country2).plot(legend = 'Upper right', color = 'blue', label = country2)
    Country_Income(country3).plot(legend = 'Upper right', color = 'black', label = country3)
    plt.legend(loc=2, prop={'size': 15})
    plt.tick_params(axis = 'both', which = 'major', labelsize = 14, length = 10)
```

```python
DF_Terror = pd.read_csv("globalterrorismdb_0718dist.csv", sep = ',', low_memory=False)
```

```python
DF_Terror.head()
```

```python
DF_Terror_merger = DF_Terror[['iyear','country_txt']].copy()
DF_Terror_merger.head()
```

```python
Yearly_Terror = DF_Terror_merger.groupby(['country_txt','iyear']).size().reset_index(name='count')
Yearly_Terror.head()
```

```python
Yearly_TerrorT = Yearly_Terror.transpose()
Yearly_TerrorT
```

```python
Yearly_TerrorT.columns = Yearly_TerrorT.iloc[0]
Yearly_TerrorT['Afghanistan']
```

```python
def Plotting_Terror(Country1, Country2, Country3):
    x = Yearly_TerrorT[Country1].iloc[[1]]
    y = Yearly_TerrorT[Country1].iloc[[2]]
    x1 = Yearly_TerrorT[Country2].iloc[[1]]
    y1 = Yearly_TerrorT[Country2].iloc[[2]]
    x2 = Yearly_TerrorT[Country3].iloc[[1]]
    y2 = Yearly_TerrorT[Country3].iloc[[2]]    
    plt.figure(figsize=(10,8))
    plt.xlabel('Year', fontsize = 18) 
    plt.ylabel('# of Terroristic Incidents', fontsize = 18) 
    
    plt.title('Terror Incidents for '+ Country1 + ", " + Country2 + " and " + Country3 , fontsize=14) 
    plt.style.use('seaborn-whitegrid')
    
    
    plt.tick_params(axis = 'both', which = 'major', labelsize = 14, length = 10)
      
    CA = plt.scatter(x,y, label = Country1, marker ='x', color='r')
    CB = plt.scatter(x1,y1, label=Country2, marker = 'x', color = 'b')
    CC = plt.scatter(x2,y2, marker = 'x', label = Country3, color='g')
    
    plt.legend((CA, CB, CC), (Country1, Country2, Country3), loc='upper left',fontsize=14)
    plt.xticks(np.arange(1970,2017,6))
```
