```python
Master_DF_2017 = pd.read_csv("NFLFantasyStats_2017.csv", header=2)
```

```python
#Takes the top 100 Fantasy point scorers #47 active players * 32 teams is 1472 players.  This is the top 0.07%
def Initial_DF100(DF):
    DF1 = DF.head(100)
    return DF1
#Renames the columns so that they are not abbreviated
def Renamed_Columns(DF):
    DF1 = Initial_DF100(DF)
    DF1.columns = ['Rank', 'Player', 'Team','Fantasy Position','Age','Games Played','Games Started as Off or Def','Passes completed','Passes attempted',"Yards Gained by Passing",'Passing Touchdowns','Interceptions Thrown','Rushing Attempts','Rusing Yards Gained','Rushing Yards per Attempt','Rushing Touchdowns','Pass Targets','Receptions','Receiving Yards','Receiving Yards per Reception','Receiving Touchdowns','# of times fumbled both lost and recovered by own team','Fumbles Lost by Player','Touchdowns of every type','Two-Point Conversions Made','Two-point Conversion Passes','Fantasy Points','Fantasy Points(PPR scoring)','Fantasy Points (Draftkings scoring)','Fantasy Points(FanDuel scoring)','Value Based Drafting','Fantasy rank within Position','Overall Fantasy Rank']
    return DF1

#Creates the skeleton for Most_Teams
def Counting_Teams(DF):
    DF_Renamed = Renamed_Columns(DF)
    DF1 = DF_Renamed[['Rank','Team']]
    return DF1

#Counts how many players are there from each team
def Most_Teams(DF):
    DF_Counted = Counting_Teams(DF)
    DF_Counts = DF_Counted.groupby(['Team'])['Rank'].count().reset_index(name = 'Count')
    return DF_Counts

#Ranks each team with the most players After doing all the previous steps
def Teams_Most_Players(DF):
    DF1 = Most_Teams(DF)
    DF2 = DF1.nlargest(10, 'Count')
    return DF2
    
```

```python
#Creating a new DF by combining previous one and the games won, Division position and year

def Won_Position_Year(DF, List_GamesWon, List_DivPos, List_Year):
    DF.insert(2, 'Games Won',List_GamesWon, True)
    DF.insert(3, 'Division Position',List_DivPos, True)
    DF.insert(4, "Year",List_Year,True )
    return DF
    
```

```python
TeamCount_2017 = Teams_Most_Players(Master_DF_2017)
```

```python
List1_2017 = [13,10,11,9,4,10,9,11,13,5]
List2_2017 = [1,3,2,2,3,1,2,1,1,4]
List3_2017 = [2017,2017,2017,2017,2017,2017,2017,2017,2017,2017]
```

```python
GDY_2017 = Won_Position_Year(TeamCount_2017, List1_2017, List2_2017, List3_2017)
```
